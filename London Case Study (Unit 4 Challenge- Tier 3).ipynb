**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
# This function calculates the percent change in Yearly Average Price for a given borough from 1995 to 2019
def create_pct_change(Borough):
    filtered_df= clean_properties[clean_properties['Borough']== Borough]
    price_2019= filtered_df[filtered_df['Year']==2019]['Average Price'].mean()
    price_1995= filtered_df[filtered_df['Year']==1995]['Average Price'].mean()
    return (price_2019/price_1995)*100

# Testing the function on Barnet 
create_pct_change('Barnet')

```

```python
# We want to create a dictionary with the percent change of each Borough
# First we create a list that houses all the borough names and converts them to strings so we can use them in a fucntion
borough_list= clean_properties['Borough'].unique().astype(str)
# Next we create an empty dict to hold out percent changes
pct_change_dict={}
# Then we loop through the boroughs, calculate pct change for each and add them to our dictionary
for borough in borough_list:
    pct_change= create_pct_change(borough)
    pct_change_dict[borough]= pct_change
#Finally we check out the dict to make sure everything looks right 
pct_change_dict
    
```

```python
# I want to create a dataframe of Pct Change to make it easier to work with 
bor= []
pct= []
# Loop through keys and append them to bor list
for borough in pct_change_dict.keys():
    bor.append(borough)
# Loop through values and append them to pct list
for value in pct_change_dict.values():
    pct.append(value)

# Create dataframe from the two lists
pct_change_df= pd.DataFrame(data=[bor, pct]).transpose()
# Rename the columns
pct_change_df.columns= ['Borough', 'Pct Change']
# Convert the Pct Change column to floats
pct_change_df['Pct Change']= pd.to_numeric(pct_change_df['Pct Change'])
# Sort by Pct Change
pct_change_df= pct_change_df.sort_values(by='Pct Change', ascending=False)
pct_change_df
```
